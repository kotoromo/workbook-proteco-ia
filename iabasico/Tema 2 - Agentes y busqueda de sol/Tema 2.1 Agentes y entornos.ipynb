# Tema 2.1: Agentes y entornos.

En el primer tema se identificó al concepto de <b>agentes racionales</b> como nuestro principal enfoque para la inteligencia artificial. En el tema II, esta noción se hace más concreta. Se verá que la racionalidad puede ser aplicada a una gran variedad de agentes que operan en cualquier entorno imaginable. Con estos conceptos se pueden desarrollar un conjunto pequeño de principios de diseño para construir agentes; sistemas que se pueden llamar <i>inteligentes</i>.

## Concepto de Agente

Un <b>agente</b> es <i>cualquier cosa que pueda percibir su <b>entorno</b> por medio de <b>sensores</b> y actuar sobre este entorno utilizando <b>actuadores</b></i>. Un agente humano tiene sensores: ojos, orejas, y otros órganos; también tiene actuadores, como lo son sus manos, piernas, etc. Un agente robótico pudiera tener cámaras y giroscopios como sensores, y partes mecánicas como sus actuadores.

El término <b>percepción</b> es utilizado para referirse a las <i>entradas de percepción en cualquier instante dado</i>. La <b>secuencia de percepciones</b> de un agente es el <i>historial completo de todo lo que el agente ha percibido</i>. Es bueno tener en mente que <i>la elección de una acción de un agente en cualquier instante dado puede depender de la secuencia de percepciones completa hasta el instante, pero no de algo que no ha sido percibido</i>.



Matemáticamente, decimos que el comportamiento de un agente es descrito por la <b>función del agente</b>, la cual mapea/asocia cualquier secuencia de percepciones a una acción. La función del agente es implementada internamente por el <b>programa del agente</b>. La función del agente es una descripción matemática abstracta mientras que el programa del agente es una implementación concreta siendo aplicada sobre un sistema físico.

Se debe tener en cuenta que la noción de agente es simplemente una herramienta para analizar sistemas; no una caracterización absoluta que divide el mundo en agentes y no agentes.

### Ejemplo Práctico: Mundo del robot aspiradora

Para ilustrar las ideas mencionadas anteriormente, vamos a utilizar un ejemplo sencillo: el de el mundo del robot aspiradora, representado por la siguiente figura:



Este mundo es tan simple que podemos describir todo lo que sucede y, como es "virtual", podemos variarlo de muchas formas. Este ejemplo en particular tiene tan sólo dos lugares: los rectángulos A y B. El agente aspiradora percibe en qué rectángulo se encuentra y si hay polvo en el lugar. Puede escoger entre moverse a la izquierda o derecha, aspirar la tierra, o hacer nada.

#### Vamos a formular una función de agente sencilla:
> Si el lugar en el que te encuentras está sucio, entonces aspira; de otra manera, ve hacía el otro rectángulo.

Una tabulación parcial de este agente se presenta a continuación.

| Secuencia de Percepciones        | Acción           |
| :------------------------------: |:----------------:|
| [A, Limpio]                        | Derecha    |
| [A, Sucio]                         | Aspirar    |
| [B, Limpio]                        | Izquierda  |
| [B, Sucio]                         | Aspirar    |
| [A, Limpio],[A, Limpio]            | Derecha    |
| [A, Limpio], [A, Sucio]            | Aspirar    |
| .                         | .    |
| .                         | .    |
| .                         | .    |
| [A, Limpio],[A, Limpio], [A, Limpio] | Derecha    |

Cabe mencionar que la tabla es una representación <i>externa</i> de la función agente que implementa de manera <i>interna</i> el programa agente. 

Al observar la tabla, apreciamos que varios agentes del mundo del robot aspiradora pueden ser definidos al llenar la columna del lado derecho en varias maneras. La pregunta entonces es: ¿Cuál es la manera correcta de llenar la tabla?, que en otras palabras es lo mísmo que preguntar ¿qué es lo que hace que un agente sea bueno, malo, inteligente o estúpido?
